```python
def convert_temperature_nan(cell):
    if cell == 'n.a':
        return 22.5
    return cell

def convert_windspeed_nan(cell):
    if cell == 'n.a':
        return 6.5
    return cell
df6 = pd.read_excel("C:\\Users\\USER\\Desktop\\Random_Data\\pandas_demo_test_excel_file.xlsx","Sheet1",converters={
    'temperature':convert_temperature_nan,
    'windspeed':convert_windspeed_nan
})
df6
```
